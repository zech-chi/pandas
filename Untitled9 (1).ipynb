```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remove_stopwords(input_string):
    # Tokenize the input string into words
    words = input_string.split()

    # Get the list of English stop words from NLTK
    stop_words = set(stopwords.words("english"))

    # Filter out the stop words from the list of words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a string
    result_string = " ".join(filtered_words)

    return result_string

import re

def split_phrase(phrase):
    # Use re.split() with a regular expression pattern
    # to split the phrase using space, comma, or period as delimiters
    words = re.split(r'[ ,.]', phrase)
    
    # Remove empty strings from the resulting list
    words = [word.strip() for word in words if word.strip() != '']
    
    return words

def all_words(list_words):
    All_words = set()
    for words in list_words:
        for word in words:
            All_words.add(word)
    return list(All_words)

def table(docs, All_words):
    res = {}
    length_docs = len(docs)
    length_all_words = len(All_words)
    dict_all_words = {}
    for i in range(length_all_words):
        dict_all_words[all_words[i]] = i

    for i in range(length_docs):
        res['d' + str(i + 1)] = [0 for _ in range(length_all_words)]
        for word in docs[i]:
            res['d' + str(i + 1)][dict_all_words[word]] += 1
    return res
```

```python
Docs = ["abdmalak", "The sky is blue", "The sun is bright today.", "The sun in the sky is bright.", "We can see the shining sun, the bright sun."]
Docs_without_empty_words = []
for d in Docs:
    Docs_without_empty_words.append(split_phrase(remove_stopwords(d)))

for d in Docs_without_empty_words:
    print(d)
```

```python
all_words = all_words(Docs_without_empty_words)
all_words
```

```python
T = table(Docs_without_empty_words, all_words)
```

```python
T
```

```python
import pandas as pd

final_table = pd.DataFrame(T).T
final_table.columns = all_words
final_table
```

```python
for k, v in T.items():
    count = 0
    for n in v:
        count += n
    for i in range(len(v)):
        T[k][i] = T[k][i] / count
print(T)
```
